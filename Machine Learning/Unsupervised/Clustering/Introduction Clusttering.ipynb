#Clusttering

```python
#Create a k-means model usig Iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertias = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

plt.plot(range(1, 11), inertias, 'bx-')
plt.title('Método del Codo')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.show()
```

```python
#Create a k-means model using iris
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
print("Etiquetas de los clusters:", labels)
print("Coordenadas de los centroides:", centroids)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.5, marker='x')
plt.title('Gráfica del Modelo KMeans con Dos Centroides')
plt.xlabel('Longitud del Sépalo (cm)')
plt.ylabel('Ancho del Sépalo (cm)')
plt.show()
```
