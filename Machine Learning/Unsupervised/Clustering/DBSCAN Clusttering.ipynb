```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Get the data
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
```

```python
iris
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
dbscan = DBSCAN(eps=0.5)
dbscan.fit(X_scaled)
```

```python
labels = dbscan.labels_
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['cluster'] = dbscan.labels_
```

```python
sns.scatterplot(x='sepal length (cm)', y='cluster', data=iris_df)
plt.title('Sepal length DBSCAN Clustering')
plt.show()
```

```python
sns.scatterplot(x='sepal width (cm)', y='cluster', color='g', data=iris_df)
plt.title('Sepal width DBSCAN Clustering')
plt.show()
```

```python
sns.scatterplot(x='petal length (cm)', y='cluster', color='y', data=iris_df)
plt.title('Petal length DBSCAN Clustering')
plt.show()
```

```python
sns.scatterplot(x='petal width (cm)', y='cluster', color='red', data=iris_df)
plt.title('Petal width DBSCAN Clustering')
plt.show()
```
